## Processing all the midi files to make training data

```python
notes=[]
for file in glob.glob('midi_songs/*.mid'):
    midi=converter.parse(file)
    print("Parsing file %s"%file)
    song=midi.flat.notes
    for el in song:
        if isinstance(el,note.Note):
            notes.append(str(el.pitch))
        elif isinstance(el,chord.Chord):
            notes.append("$".join(str(n) for n in el.normalOrder))
    
```

## saving the training data to save time

```python
with open('notes','wb') as filepath:
    pickle.dump(notes,filepath)
```

```python
with open('notes','rb') as fp:
    notes=pickle.load(fp)
```

```python
distinct_nodes_chords=set(notes)
print('Total notes and chords -',len(notes))
print('Distinct notes and chords -',len(distinct_nodes_chords))
```

## Sequential data for LSTM
##### as the music is a sequential time varying data LSTM is used

```python
# Input layer parameters
sequence_len=100
```

```python
classes=sorted(distinct_nodes_chords)
```

```python
# mapping these distinct classes to a numerical value
pitch_map=dict((element,ind) for ind,element in enumerate(classes))
```

```python
network_input=[]
network_output=[]
```

```python
for i in range(len(notes)-sequence_len):
    i_input=notes[i:i+sequence_len]     # 100 inputs
    i_out=notes[i+sequence_len]
    
    network_input.append([pitch_map[entry] for entry in i_input])
    network_output.append(pitch_map[i_out])
```

```python
len(notes)-sequence_len
```

```python
len(network_input)
```

```python
# no of sample points
n_points=len(network_input)
```

```python
network_input=np.reshape(network_input,(n_points,sequence_len,1))
# 1 for converting it into 3d
```

```python
network_input.shape
```

```python
## max scale normalising of the data
normalized_network_input=network_input/float(len(distinct_nodes_chords))
```

```python
## encoding of the output classes 
```

```python

from keras.utils import np_utils
```

```python
network_output=np_utils.to_categorical(network_output)
```

```python
network_output[1]
```

# Creating Model Architecture

```python
from keras.models import Sequential,load_model
from keras.layers import *
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
!conda install numpy==1.18.5
```

```python
print(normalized_network_input.shape[1],normalized_network_input.shape[2])
```

```python
model=Sequential()
model.add(LSTM(units=512,
               input_shape=(normalized_network_input.shape[1],normalized_network_input.shape[2]),
               return_sequences=True ))
model.add(Dropout(0.3))
model.add(LSTM(units=512,return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(len(distinct_nodes_chords),activation='softmax'))
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam")
```

```python
model.summary()
```

```python
checkpoint=ModelCheckpoint("model.hdf5",monitor="loss",verbose=0,save_best_only=True,mode="min")

model_hist=model.fit(normalized_network_input,network_output,epochs=100,batch_size=64,callbacks=[checkpoint])
```

```python
model=load_model('weights.hdf5')
```

### Predicting new sequences

```python
network_input=[]
for i in range(len(notes)-sequence_len):
    i_input=notes[i:i+sequence_len]     # 100 inputs
    
    network_input.append([pitch_map[entry] for entry in i_input])
```

```python
network_input[300]
```

```python

# mapping int to music element
```

```python
start=np.random.randint(len(network_input)-1)
int_to_ele=dict((num,ele) for num,ele in enumerate(distinct_nodes_chords))
# initial pattern
pattern=network_input[start]
prediction_output=[]

#generate 200 elements
for note_ind in range(120):
    prediction_input=np.reshape(pattern,(1,len(pattern),1))
    prediction_input=prediction_input/len(distinct_nodes_chords)
    
    prediction=model.predict(prediction_input,verbose=0)
    index=np.argmax(prediction)
    res=int_to_ele[index]
    prediction_output.append(res)
    pattern.append(index)
    pattern=pattern[1:]
    
```

```python
print(prediction_output)
```

### Create MIDI from prediction

```python
offset=0 ## time
output_notes=[]

for el in prediction_output:
    # if it is a chord
    if ('$' in el) or el.isdigit():
        notes_in_chord=el.split('$')
        tmp_notes=[]
        for n in notes_in_chord:
            new_note=note.Note(int(n))
            new_note.storedInstrument=instrument.Guitar()
            tmp_notes.append(new_note)
        chord_obj=chord.Chord(tmp_notes)
        chord_obj.offset=offset
        output_notes.append(chord_obj)
    
    
    
    # if note
    else:
        new_note=note.Note(el)
        new_note.offsetfset=offset
        new_note.storedInstrument=instrument.Guitar()
        output_notes.append(new_note)
    offset+=0.5
```

```python
# stream object to play the generated file
midi_stream=stream.Stream(output_notes)
midi_stream.write('midi',fp='output.mid')
```
